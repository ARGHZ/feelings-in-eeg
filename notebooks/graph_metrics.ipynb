```python
%matplotlib inline
import config as general_config
import json

import networkx as nx
from networkx.readwrite import json_graph
from digraph import draw_graph

from featuresextractor import get_graph_metrics_table, get_csv_content

subj, emotion = 'subj_1', 'sad'
```

```python
graph_metrics_all = get_csv_content(general_config.DATA_DIR + '/graph_metrics_all.csv')
```

```python
freq_band = 'delta'
file_name = subj + '_' + emotion + '_raw_try-1.mat.json'
json_file_path = general_config.DATA_DIR + '/graphs/' + freq_band +'/' + file_name
with open(json_file_path) as json_file:
    data = json.load(json_file)

DG = json_graph.node_link_graph(data)

draw_graph(DG, '')
```
